```python
import mdtraj
import numpy as np
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1,1, figsize=(15,5))
for ens in analysis.trajectories:
    res_based_sasa = mdtraj.shrake_rupley(analysis.trajectories[ens], mode='residue')
    sasa_mean = np.mean(res_based_sasa, axis=0)
    sasa_std = np.std(res_based_sasa, axis=0)

    plt.plot(np.mean(res_based_sasa, axis=0),linestyle= '--' , marker='o' ,  markersize=3)
    plt.fill_between([i for i in range(res_based_sasa.shape[1])], sasa_mean - sasa_std , sasa_mean + sasa_std,interpolate=True ,alpha = 0.35)
plt.show()
```
